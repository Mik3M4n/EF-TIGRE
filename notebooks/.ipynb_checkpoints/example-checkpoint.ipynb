# Example

Here we provide an example of how to use the package. We set the values of the free parameters, solve for the background evolution, compute the perturbations and the signal

## Setup

```python
import os
import sys
import numpy as np
import matplotlib.pyplot as plt



PACKAGE_PARENT = '..'
SCRIPT_DIR = os.path.dirname(os.path.realpath(os.path.join(os.getcwd())))
sys.path.append(SCRIPT_DIR)
print(SCRIPT_DIR)
```

```python
# From Planck 18, https://arxiv.org/pdf/1807.06209.pdf TT,TE,EE+lowE+lensing+BAO table 2 last column

THETA_STAR =  1.04101/100
R_STAR =  144.57 #Mpc
DM_LS = R_STAR/THETA_STAR

Z_STAR = 1089.80
X_STAR = -np.log(1+Z_STAR)

Ombh2 = 0.02242
clight = 299792.458

H0=h*100

```

```python
# Set here the value of h = H0/100. 
# Planck: 0.6766
# Sne: 0.7304

h = 0.6766 
```

```python
# ... then, the following quantities are changed:

Om_b =  Ombh2/( (h)**2)
Omr0 =  4.18343*1e-05/h**2
H0=h*100

# dimensioneless comoving distance to last scattering (Mpc)
DM_LS_DIMLS = DM_LS*H0/clight
```

```python
settings = { "zin": 1100,
            "zin_pert":10,
            "tolBG": 1e-8,
            "tolPert": 1e-05,
            "Om0Pl": 0.3111,
            "Ob0Pl": Om_b,
            "H0": H0,
            

}

# We use x = log(a) as time variable. These are the initial values
xin = -np.log(1+settings['zin'])
xin_cs = -np.log(1+settings['zin_pert'])
```

```python
# set here the values of the MG parameters in different fiducial models


MGparams = { 'aM' : { 'LCDM': 0., 'MGI': -0.1},       
             'aB' : { 'LCDM': 0., 'MGI': -0.5},
             'gammaC' : { 'LCDM': 0., 'MGI': 0.15},
             'w' : { 'LCDM': -1., 'MGI': -1.41}

}


fnames = ['LCDM', 'MGI']
```

## Background

```python
import likelihood_tools as lt
```

```python

all_BGs = {}

for fiducial in fnames:

    print('*'*80)
    print('Fiducial: %s\n' %fiducial)

    aM = MGparams['aM'][fiducial]
    aB = MGparams['aB'][fiducial]
    gamma_par = MGparams['gammaC'][fiducial]
    w = MGparams['w'][fiducial]
    
    my_BG = lt.solve_bg(aM, aB, gamma_par, w, settings, verbose=True, is_beta_gamma=False, H_int='backward')

    all_BGs[fiducial] = my_BG
    
    print('Done.\n')
```

### Plot bg quantities

```python
xgrid = np.geomspace(xin, -1e-5, 1000)
zgrid = -1+np.exp(-xgrid)
```

```python
# Energy density evolution

clist = ['gold', 'blueviolet', 'navy', 'firebrick', 'turquoise']
lss = ['-', ':']

for i,fiducial in enumerate(fnames):

    
    plt.plot(zgrid, all_BGs[fiducial].Ob(xgrid), label='Baryons '+fiducial, color = clist[0], ls=lss[i] )
    plt.plot(zgrid, all_BGs[fiducial].Oc(xgrid), label='CDM '+fiducial, color = clist[1], ls=lss[i] )
    plt.plot(zgrid, all_BGs[fiducial].Om(xgrid), label='Total matter '+fiducial,  color = clist[2], ls=lss[i] )
    plt.plot(zgrid, all_BGs[fiducial].ODE(xgrid), label='Dark energy '+fiducial, color = clist[3], ls=lss[i] )
    plt.plot(zgrid, all_BGs[fiducial].Or(xgrid), label='Radiation '+fiducial, color = clist[4], ls=lss[i] )



plt.xscale('log')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.title(r'Energy fraction $\Omega$', fontsize=15)
plt.xlabel(r'Redshift $z$', fontsize=15)
plt.ylabel(r'$\Omega$', fontsize=15)

plt.show()
```

```python
# Sound speed

for fiducial in fnames:

    plt.plot(zgrid, all_BGs[fiducial].cs2al(xgrid), label=fiducial,  lw=3,  )

plt.xscale('log')
plt.yscale('log')
plt.xlim( 1e-05, 100  )
plt.legend()
plt.title('Sound speed', fontsize=15)
plt.xlabel(r'Redshift $z$', fontsize=15)
plt.ylabel(r'$c_s^2 \times \alpha$', fontsize=15)
plt.show()
```

## Perturbations

```python
from perturbations import Perturbations
```

```python
X = 1.
```

```python

all_pert = {}

for fiducial in fnames:

    print('*'*80)
    print('Fiducial: %s\n' %fiducial)


    myPert = Perturbations(all_BGs[fiducial], zin=settings['zin_pert'], tol=1e-15, X=X)
    
    myPert.solve(method='DOP853', res=100, rtol=settings['tolPert'], atol=settings['tolPert']**2, verbose=True) 

    all_pert[fiducial] = myPert

    print('Done.\n')
```

### Plots

```python
for i,fiducial in enumerate(fnames):

    
    plt.plot(zgrid, all_pert[fiducial].f_eff(xgrid), label=fiducial)
    


plt.xscale('log')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.title(r'Effective growth rate ', fontsize=15)
plt.xlabel(r'Redshift $z$', fontsize=15)
plt.ylabel(r'$f_{\rm eff}$', fontsize=15)

plt.show()
```

## Multipoles

```python
from multipole_signal import Signal
```

```python
# put here which multipoles you want to compute

which_multipoles = ["dipole"] #["monopole", "dipole", "quadrupole", "hexadecapole"]
```

```python
b1B, b1F = 0.554, 0.554
b2B, b2F = 0.783, 0.783

sB_fit_1 = 0.953
sB_fit_2 = 0.691
sB_fit_3 = 0.18

sF_fit_1 = -0.105
sF_fit_2 = -0.01
sF_fit_3 = -0.029
```

```python
from mul_nu1 import interpolate_mul_nu1
mul_nu1=interpolate_mul_nu1(base_dir='..')
```

```python


all_signals = {}

for fiducial in fnames:

    print('*'*80)
    print('Fiducial: %s\n' %fiducial)


   
    mySignal = Signal(all_BGs[fiducial], 
                      all_pert[fiducial], 
                      mul_nu1, 
                      X=X, 
                      H0=1/2997.9, # this is just a dimensionful constant, not the actual H0
                      b1 = [b1B, b1F], 
                      b2 = [b2B, b2F], 
                      sB_fit = [sB_fit_1, sB_fit_2, sB_fit_3], 
                      sF_fit = [sF_fit_1, sF_fit_2, sF_fit_3], 
                      which_multipoles= which_multipoles, 
                      zin=settings["zin_pert"], 
                      dipole_boost=1. )


    all_signals[fiducial] = mySignal
    
    print('Done.\n')
```

### Plots

```python
z = 0.15
d = np.linspace(20, 200)
```

```python
for fiducial in fnames:

    svals = np.asarray([all_signals[fiducial].calculate_signal(d_,z) for d_ in d])

    plt.plot(d, svals, label=fiducial)

plt.xscale('log')
plt.yscale('log')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.xlabel(r'Distance $d$', fontsize=15)
plt.ylabel(r'Dipole', fontsize=15)

plt.show()
```
